```python
import pandas as pd
import time

# 1. Start a timer to monitor the loading speed
print("Loading the entire dataset... Please wait (this can take up to a minute).")
start_time = time.time()

# 2. Path to your specific file layout
file_path = r"C:\Users\harep\Downloads\01032026-20260616T103430Z-3-001\01032026\12042026\Online_Retail.xlsx"

# 3. Read the entire file (no nrows limit)
df = pd.read_excel(file_path)

# 4. Calculate total load time and display shape
end_time = time.time()
print(f"Success! Data loaded in {end_time - start_time:.2f} seconds.")
print("Total rows and columns found:", df.shape)

# 5. Show the first 5 rows of the full dataset
df.head()
```
