# Decorators (Advanced)

```python
# Decorators (Advanced)

# This is the decorator function. 
# It takes another function (func) as an argument — that’s key because decorators “wrap” other functions.

# Functions with no parameters
def decorator_example(func):
    def wrapper():
        print("Before the function call")
        func()
        print("After the function call")
    return wrapper

@decorator_example
def say_hello():
    print("Hello!")
    
@decorator_example
def say_WhatUrdoing():
    print("WhatUrdoing!")


# The @decorator_example applies the decorator to say_hello(). 
# So when you call say_hello(), Python actually calls wrapper() from the decorator.

say_hello()
say_WhatUrdoing()
```

```python
#works for any number of arguments

def decorator_example(func):
    def wrapper(*args, **kwargs):
        print("Before the function call")
        result = func(*args, **kwargs)
        print("After the function call")
        return result
    return wrapper


@decorator_example
def greet(name,age):
    print(f"Hello, {name}!, {age}")


@decorator_example
def wishing(Quote,Who):
    print(f"QoD, {Quote}!, {Who}")


greet("Alice","25")

wishing("If you chase butterflies, they'll fly away. But if you make a beautiful garden, the butterflies will come to you", "German Saying")
```

```python
def repeat(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):
                func(*args, **kwargs)
        return wrapper
    return decorator


@repeat(3)
def say_hi():
    print("Hi!")

say_hi()

# repeat(3) → returns decorator
# decorator(func) → returns wrapper
# wrapper() → runs the function
```

```python
#inbuilt wrap function -> functools.wraps
from functools import wraps

def decorator_example(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result
    return wrapper


@repeat(3)
def say_hi():
    print("Hi!")

say_hi()
```

```python
import time

#use case. You can measure latency/time taken of any function through this

def timing_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)# the actual function where it being called
        end_time = time.time()
        print(f"Execution time: {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@timing_decorator
def compute_sum(n):
    return sum(range(n))

# Compute the sum of the first million numbers
print(compute_sum(1000000))
```

```python
# Imagine you're building an API and only logged-in users can access certain functions.
from functools import wraps

# Simulated logged-in user
current_user = {"username": "Alice", "authenticated": True}

def login_required(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if not current_user["authenticated"]:
            print("Access denied. Please log in.")
            return
        return func(*args, **kwargs)
    return wrapper


@login_required
def view_dashboard():
    print("Welcome to your dashboard!")

view_dashboard()

# Frameworks like Flask and Django use this pattern heavily.
# Example in real frameworks
#     @login_required 
#     @permission_required  
#     @admin_only
```

```python
from functools import wraps

def simple_cache(func):
    cache = {}

    @wraps(func)
    def wrapper(*args):
        if args in cache:
            print("Returning cached result...")
            return cache[args]

        result = func(*args)
        cache[args] = result
        return result

    return wrapper


@simple_cache
def expensive_calculation(n):
    print("Performing expensive calculation...")
    return n * n


print(expensive_calculation(5))
print(expensive_calculation(5))  # This time it's cached


Used in:

    # APIs
    # Machine learning pipelines
    # Data processing

# inbuilt function - from functools import lru_cache
```

```python
from functools import wraps
import datetime

def log_execution(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"[{datetime.datetime.now()}] Running {func.__name__}")
        result = func(*args, **kwargs)
        print(f"[{datetime.datetime.now()}] Finished {func.__name__}")
        return result
    return wrapper


@log_execution
def process_payment(amount):
    print(f"Processing payment of ${amount}")

process_payment(100)

# Used in:

# Payment systems
    # Banking apps
    # Backend services
    # Microservices logging
```

```python
# Higher-Order Functions: map, filter, reduce
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map: Apply a function to each element
squares = list(map(lambda x: x ** 2, numbers))
print(f"Squares: {squares}")

# filter: Keep elements that match a condition
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers: {even_numbers}")

# reduce: Combine elements into a single result
product = reduce(lambda x, y: x * y, numbers)
print(f"Product of numbers: {product}")
```

```python
# Error Handling in Functions

#print(5/0)# Run this First

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero!"
        
print(safe_divide(10, 3))
print(safe_divide(5, 0))
```
